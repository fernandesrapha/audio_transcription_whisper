```python
# ✅ 3. Importar o Whisper e carregar o modelo
import whisper

# Carrega o modelo
model = whisper.load_model("turbo")  # você pode trocar por "small", "medium", "large-v3", "turbo" para mais qualidade
```

```python
# ✅ 2. Fazer upload dos arquivos de áudio
from google.colab import files
import os

uploaded = files.upload()

# ✅ 4 e 5. Iterar sobre todos os arquivos enviados, transcrever e salvar
for audio_path in uploaded.keys():
    print(f"\nTranscrevendo: {audio_path}")

    # Transcreve o áudio com idioma definido como português
    result = model.transcribe(audio_path, language='pt')
    transcription_text = result["text"]

    # Cria o nome do arquivo .txt baseado no nome do arquivo de áudio original
    file_name = os.path.splitext(audio_path)[0] + ".txt"

    # Salva a transcrição no arquivo .txt com codificação UTF-8
    with open(file_name, "w", encoding="utf-8") as f:
        f.write(transcription_text)

    print(f"Transcrição salva em: {file_name}")

    # Executa o download automático do arquivo gerado
    files.download(file_name)
```
